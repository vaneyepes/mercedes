## Evaluation of Ridge regression with grid search

**Best parameters: {'alpha': 100.0}**

- Out of the grid [0.001, 0.01, 0.1, 1, 10, 100], the best performing alpha was 100.

- High alpha = strong regularization - stronger penalty (control overfitting / multicollinearity for a high-dimensional dataset (551 features))

**Best CV R²: 0.510**

- Ridge explained ~51% of the variance on data (good generalization)
 
**Test Performance**

**MAE: 5.592** On average, predictions are off by about 5.6 seconds (Note: MAE is easier to interpret than RMSE)

**RMSE: 8.221** average error, giving more weight to larger mistakes, is 8.2 seconds

**R²: 0.566** The model explains ~57% of the variance in y

- Better than Top-10 Linear Regression (R² = 0.557) and much better than Top-3 Linear (R² = 0.403).

- Ridge iprove predictive power a little but is preventing overfitting.


## Conclusions Ridge regression 

- Ridge with all features gave the best performance so far:

- RMSE improved to 8.2 vs 9.6 (Top-3) and 8.3 (Top-10)

- R² improved to 0.566, explaining more variance than simpler baselines

- Regularization helped the model make use of all features without overfitting

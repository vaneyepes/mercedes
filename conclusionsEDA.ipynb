# Conclusions Including encoded categorical features


## Correlation Analysis with Encoded Features

- After applying **One-Hot Encoding (OHE)** to the categorical variables, some encoded features appear correlated with the target `y`.  
- For model building, we ranked features by **absolute correlation** (|corr|), since both positive and negative relationships can be predictive in regression.  

### Key Findings
- **Strongest positive correlations with `y`:**
  - `X314` (+0.61), `X261` (+0.59), `X263` (+0.38), `X136` (+0.37).  
- **Strongest negative correlations with `y`:**
  - `X127` (−0.51), `X279` (−0.38), `X232` (−0.38), `X29` (−0.38), plus several others in the −0.36 to −0.37 range.  
- **Moderate OHE features:**
  - Categories such as `X0_ak` (+0.28) and `X0_x` (+0.26) show moderate positive correlations, but were not as strong as the features above.  

### Interpretation
- Both **positive and negative correlations** are informative predictors.  
- For example, higher values of `X314` and `X261` are associated with higher test times, while higher values of `X127` are associated with lower test times.  
- Certain categories from categorical features (e.g., `X0_ak`) also influence test times, but less strongly.  
- Several constant columns (`X107`, `X233`, `X235`, `X268`, `X289`, `X290`, `X293`, `X297`, `X330`, `X347`) returned `NaN` because they never vary (low variance)→ these were dropped as uninformative.
